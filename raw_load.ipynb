```python
%sql

-- Create Catalog
CREATE CATALOG IF NOT EXISTS MOVIELENS;


-- create raw schema for initial loading
CREATE SCHEMA IF NOT EXISTS MOVIELENS.RAW;
```

```python

ACCESS_KEY = "PASTE_YOUR_AWS_ACCESS_KEY_ID_HERE"
SECRET_KEY = "PASTE_YOUR_AWS_SECRET_ACCESS_KEY_HERE"


# ====== CONFIGURE SPARK TO READ FROM S3 ======
hconf = spark._jsc.hadoopConfiguration()
hconf.set("fs.s3a.access.key", ACCESS_KEY)
hconf.set("fs.s3a.secret.key", SECRET_KEY)
hconf.set("fs.s3a.endpoint", "s3.amazonaws.com")

# Optional but often helpful
hconf.set("fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
hconf.set("fs.s3a.path.style.access", "true")

# ====== READ ALL CSV FILES FROM YOUR BUCKET ======
df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("s3a://saimovieliens/*.csv")
)

display(df)
```
